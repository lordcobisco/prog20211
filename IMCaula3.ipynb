```python
nome = input("Digite seu nome: ")
peso = float(input("Digite seu peso: "))
altura = float(input("Digite sua altura em metros: "))
IMC = peso/(altura**2)

print(nome, ', seu imc é: ', IMC)

print('abaixo do peso', IMC <17)
print('peso normal', 17.0 < IMC <18.5)
print('peso dentro do normal', 18.5 < IMC < 25.0)
print('Acima do peso', 25.0 < IMC < 30.0)
print('Muito acima do peso', IMC >30.0)


```
